1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp

x = sp.symbols('x')

function = 10 * x - 5
derivative_function = sp.diff(function, x)
```

```python
derivative_function
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show();
```

Кожна додаткова година навчання в 10 разів збільшує можливість отримати кращу оцінку.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x1 = sp.symbols('x1')

function_1 = sp.sin(x1)
derivative_function_1 = sp.diff(function_1, x1)
```

```python
print(function_1)
print(derivative_function_1)
```

```python
x2 = sp.symbols('x2')

function_2 = sp.ln(x2)
derivative_function_2 = sp.diff(function_2, x2)
```

```python
print(function_2)
print(derivative_function_2)
```

```python
plt.figure(figsize=(14, 10))

x1 = np.linspace(0, 10, 100)
f1 = np.sin(x1)
df1 = np.cos(x1)

plt.subplot(2, 1, 1)
plt.plot(x1, f1, label='f(x) = sin(x)', color='blue')
plt.plot(x1, df1, label="f'(x) = cos(x)", color='green')
plt.title("Графік функції та її похідної для sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

x2 = np.linspace(0.1, 10, 100)
f2 = np.log(x2)
df2 = 1 / x2

plt.subplot(2, 1, 2)
plt.plot(x2, f2, label='f(x) = log(x)', color='blue')
plt.plot(x2, df2, label="f'(x) = 1/x", color='green')
plt.title("Графік функції та її похідної для ln(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show();
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

function_3 = x**2 - 5 * x + 1
derivative_function_3 = sp.diff(function_3, x)

print(function_3)
print(derivative_function_3)
```

```python
x = np.linspace(0, 100, 1000)
f = x**2 - 5*x + 1
df = np.full_like(f, x)

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = x² - 5х + 1', color='blue')
plt.plot(x, df, label="f'(x) = 2x - 5", color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show();
```
